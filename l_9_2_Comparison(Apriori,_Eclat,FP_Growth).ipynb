```python
import pandas as pd
import time


data = pd.read_csv('Market_Basket_Optimisation.csv', header=None)
# Convert to transaction list
transactions = []
for i in range(len(data)):
    transactions.append([str(data.values[i, j])
                         for j in range(data.shape[1])
                         if str(data.values[i, j]) != 'nan'])
```

```python
#Apriori
from mlxtend.preprocessing import TransactionEncoder
from mlxtend.frequent_patterns import apriori

te = TransactionEncoder()
te_array = te.fit(transactions).transform(transactions)
df = pd.DataFrame(te_array, columns=te.columns_)

start = time.time()
apriori_result = apriori(df, min_support=0.003, use_colnames=True)
apriori_time = time.time() - start

print("Apriori Itemsets:", len(apriori_result))
print("Apriori Time:", apriori_time)
```

```python
#Apriori
from mlxtend.preprocessing import TransactionEncoder
from mlxtend.frequent_patterns import apriori

te = TransactionEncoder()
te_array = te.fit(transactions).transform(transactions)
df = pd.DataFrame(te_array, columns=te.columns_)

start = time.time()
apriori_result = apriori(df, min_support=0.003, use_colnames=True)
apriori_time = time.time() - start

print("Apriori Itemsets:", len(apriori_result))
print("Apriori Time:", apriori_time)
```

```python
#Eclat
from collections import defaultdict

start = time.time()

item_tid = defaultdict(set)
for tid, transaction in enumerate(transactions):
    for item in transaction:
        item_tid[item].add(tid)

min_support = 0.003 * len(transactions)
eclat_result = {item: tids for item, tids in item_tid.items()
                if len(tids) >= min_support}

eclat_time = time.time() - start

print("Eclat Itemsets:", len(eclat_result))
print("Eclat Time:", eclat_time)
```

```python
#FP-Growth
from mlxtend.frequent_patterns import fpgrowth

start = time.time()
fpgrowth_result = fpgrowth(df, min_support=0.003, use_colnames=True)
fpgrowth_time = time.time() - start

print("FP-Growth Itemsets:", len(fpgrowth_result))
print("FP-Growth Time:", fpgrowth_time)
```

```python
comparison = pd.DataFrame({
    'Algorithm': ['Apriori', 'Eclat', 'FP-Growth'],
    'Frequent Itemsets': [
        len(apriori_result),
        len(eclat_result),
        len(fpgrowth_result)
    ],
    'Execution Time (sec)': [
        apriori_time,
        eclat_time,
        fpgrowth_time
    ]
})

print(comparison)
```
