```python
import pandas as pd
from mlxtend.preprocessing import TransactionEncoder
from mlxtend.frequent_patterns import apriori, association_rules
```

```python
import pandas as pd
data = pd.read_csv('Market_Basket_Optimisation.csv')
data.head()  
```

```python
print("Shape of dataset:", data.shape)

print("\nFirst 5 rows:")
print(data.head())

print("\nMissing values per column:")
print(data.isnull().sum())

print("\nTotal transactions:", len(data))
```

```python

data = data.fillna('')

for col in data.columns:
    data[col] = data[col].astype(str).str.strip().str.lower()
```

```python
transactions = []

for i in range(len(data)):
    items = [item for item in data.iloc[i] if item != '']
    transactions.append(items)

print("\nSample cleaned transaction:", transactions[0])



te = TransactionEncoder()
encoded = te.fit(transactions).transform(transactions)
df = pd.DataFrame(encoded, columns=te.columns_)
```

```python

freq_items = apriori(df, min_support=0.02, use_colnames=True)

print("\nFrequent Itemsets:")
print(freq_items.head())
```

```python
rules = association_rules(freq_items, metric='confidence', min_threshold=0.3)

print("\nAssociation Rules:")
print(rules[['antecedents','consequents','support','confidence','lift']].head())
```
